```python
import os
import pandas as pd
from zipfile import ZipFile
import numpy as np


```

# **Separating Dimensions**

```python
zip_files = ['/content/Data-All Algos (3).zip', '/content/Actual Values (3).zip']
extract_dirs = [f'extracted_{i+1}' for i in range(len(zip_files))]
output_dirs = [f'separated_{i+1}' for i in range(len(zip_files))]
```

```python
for output_dir in output_dirs:
    os.makedirs(output_dir, exist_ok=True)
```

```python
def process_zip(zip_file, extract_dir, output_dir):
    os.makedirs(extract_dir, exist_ok=True)
    with ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print(f"Extracted {zip_file} to {extract_dir}")


    for root, _, files in os.walk(extract_dir):
        for file in files:
            file_path = os.path.join(root, file)

            if file.endswith('.xlsx'):
                df = pd.read_excel(file_path)
            elif file.endswith('.txt'):
                df = pd.read_csv(file_path, delimiter='\t')
            else:
                continue
 # Separate 10D and 20D
            if '10D' in file:
                output_path = os.path.join(output_dir, f"10D_{file}.xlsx")
                df.to_excel(output_path, index=False)
                print(f"Saved 10D data to {output_path}")
            elif '20D' in file:
                output_path = os.path.join(output_dir, f"20D_{file}.xlsx")
                df.to_excel(output_path, index=False)
                print(f"Saved 20D data to {output_path}")

for zip_file, extract_dir, output_dir in zip(zip_files, extract_dirs, output_dirs):
    process_zip(zip_file, extract_dir, output_dir)
```

# **mean errors and std**

```python
separated_1 = ['/content/separated_1/10D_Results_for_10D.txt.xlsx' ,'/content/separated_1/20D_Results_for_20D.txt.xlsx']
separated_2 = ['/content/separated_2/10D_Results_for_10D.txt.xlsx' ,'/content/separated_2/20D_Results_for_20D.txt.xlsx']
```

```python
output_dir = '/content/output'
os.makedirs(output_dir, exist_ok=True)

results_df = pd.DataFrame(columns=['Dimension', 'Obtained File', 'Optimal File', 'Mean Error', 'Standard Deviation'])

for obtained_path, optimal_path in zip(separated_1, separated_2):
    if '10D' in obtained_path:
        dimension = '10D'
    else:
        dimension = '20D'
        continue

    if '10D' in optimal_path:
        dimension = '10D'
    else:
        dimension = '20D'
        continue

    try:
        obtained_df = pd.read_excel(obtained_path)
        optimal_df = pd.read_excel(optimal_path)
    except Exception as e:
        print(f"Error reading one of the files: {e}")
        continue
    try:
        mean_error = np.mean(obtained_df.iloc[:, 1] - optimal_df.iloc[:, 1])
        std_dev = np.std(obtained_df.iloc[:, 1])
    except Exception as e:
        print(f"Error performing calculations on files: {e}")
        continue

    temp_df = pd.DataFrame({
        'Dimension': [dimension],
        'Obtained File': [os.path.basename(obtained_path)],
        'Optimal File': [os.path.basename(optimal_path)],
        'Mean Error': [mean_error],
        'Standard Deviation': [std_dev]
    })
    results_df = pd.concat([results_df, temp_df], ignore_index=True)


output_file = os.path.join(output_dir, 'errors_statistics.xlsx')
results_df.to_excel(output_file, index=False)
print(f"Results saved to {output_file}")
```
